```python
from googlesearch import search

def search_airnav_for_muskegon():
    # Define the query to search for 'Muskegon' on the AirNav website
    query = "MUSKEGON airnav site:airnav.com"
    
    # Perform the search and get URLs
    search_results = search(query, num_results=5)  # Limiting to 5 results for simplicity
    
    print("Search results for 'Muskegon' on AirNav:")
    for result in search_results:
        print(result)

# Example usage
search_airnav_for_muskegon()
```

```python
import requests
from bs4 import BeautifulSoup

# Define the URL for Muskegon County Airport (KMKG) on AirNav
url = "https://www.airnav.com/airport/KMKG"

# Set headers to mimic a real browser request
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.36"
}

# Send the request to the webpage
response = requests.get(url, headers=headers)

# Check if the request was successful
if response.status_code == 200:
    # Parse the HTML content with BeautifulSoup
    soup = BeautifulSoup(response.content, 'html.parser')
    
    # Extract the page title to confirm correct page
    page_title = soup.title.string
    print(f"Page Title: {page_title}")
    
    # Print out the HTML structure to understand how the information is organized
    # This will help in identifying the right tags (Uncomment if needed)
    # print(soup.prettify())
    
    # Look for all <td> tags, as this is a common tag for FAA information on AirNav
    text_blocks = soup.find_all('td')
    
    # Loop through the text blocks to find specific information
    for block in text_blocks:
        if 'FAA Identifier' in block.text:
            print("FAA Identifier Info:", block.text.strip())
        if 'Lat/Long' in block.text:
            print("Lat/Long Info:", block.text.strip())
        if 'Elevation' in block.text:
            print("Elevation Info:", block.text.strip())
        if 'Variation' in block.text:
            print("Variation Info:", block.text.strip())
        if 'From city' in block.text:
            print("From City Info:", block.text.strip())
        if 'Time zone' in block.text:
            print("Time Zone Info:", block.text.strip())
else:
    print(f"Failed to fetch the page, status code: {response.status_code}")
```
